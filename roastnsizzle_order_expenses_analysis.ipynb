```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
orders = pd.read_csv('order-placement-form (Responses) - Form Responses 1.csv')

expenses = pd.read_csv('expense form (Responses) - Form Responses 1.csv')

orders.head(20)
```

```python
expenses.head(20)
```

```python
orders.dtypes
```

```python
expenses.dtypes
```

```python
orders1 = orders.drop(['timestamp'],axis = 1)
expenses1 = expenses.drop(['Timestamp','Input expense description'],axis = 1)
```

```python
orders1['customer number'] = orders1['customer number'].astype('str')
```

```python
orders1.dtypes
```

## orders analysis

```python
plt.figure(figsize=(25,10))

orders1.groupby('customer number')['Total spent'].sum().sort_values(ascending=False).plot.bar()
```

```python
plt.figure(figsize=(25,10))

orders1.groupby('customer number')['customer number'].count().sort_values(ascending = False).plot.bar()

# sortedorders.head(30)
```

```python
orders1['date ordered'] = pd.to_datetime(orders1['date ordered'])
```

```python
orders1.head(15)
```

```python
month1ord = orders1[orders1['date ordered'] <= '2021-10-10']
month2ord = orders1[orders1['date ordered'] >= '2021-10-11']

month1ord.shape
```

```python
month1sorted = month1ord.groupby('customer number')['customer number'].count().sort_values(ascending = False)

month1sorted.head(10)
```

```python
month1totalord = month1ord['Total spent'].sum()
```

```python
month2ord.shape
```

```python
month2sorted = month2ord.groupby('customer number')['customer number'].count().sort_values(ascending = False)

month2sorted.head(10)
```

```python
month2totalord = month2ord['Total spent'].sum()
```

## Expenses analysis

```python
expenses1['Input expense date']  = pd.to_datetime(expenses['Input expense date'])
```

```python
month1exp = expenses1[expenses1['Input expense date'] <= '2021-10-10']
month2exp = expenses1[expenses1['Input expense date'] >= '2021-10-11']

month1exp.shape
```

```python
month2exp.shape
```

```python
exp1sorted = month1exp.groupby('Input expense item')['Input expense amount'].sum().sort_values(ascending = False)

exp1sorted.head(20)

## sorted expense values for sept 10 -oct 10 in descending order,displaying the top 20 expense items
```

```python
exp2sorted = month2exp.groupby('Input expense item')['Input expense amount'].sum().sort_values(ascending = False)

exp2sorted.head(20)


## sorted expense values for oct 10 - nov 10 in descending order,displaying the top 20 expense items
```

```python
month1totalsexp = month1exp['Input expense amount'].sum()
```

```python
month2totalsexp = month2exp['Input expense amount'].sum()
```

## combined analysis

```python
from PIL import Image
df = pd.DataFrame([['sept',month1totalord,month1totalsexp],['oct',month2totalord,month2totalsexp]])

df.columns = ['months','sales','expenses']
# df.set_index(['months'],inplace = True)

x = np.arange(len(df.months))
width = 0.35

fig, ax = plt.subplots(figsize= (20,10))
rects1 = ax.bar(x - width/2, df.sales, width, label=df.months)
rects2 = ax.bar(x + width/2, df.expenses, width, label=df.months)

ax.set_ylabel('KES')
ax.set_title('comparison of sales vs expenses for sept/oct')
ax.set_xticks(x, df.months)
ax.legend()

ax.bar_label(rects1, padding=3)
ax.bar_label(rects2, padding=3)

fig.tight_layout()

plt.savefig('salesvexpesnes.png')

im = Image.open('salesvexpesnes.png')
rgb_im = im.convert('RGB')
rgb_im.save('salesvexpesnes.jpg','JPEG')
```
